What was the best month for sales? How much was earned that month?

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype(int)
df['Price Each'] = df['Price Each'].astype(float)

df['Total Sales'] =df['Quantity Ordered']* df['Price Each']
```

What products are most often sold together?

```python
sold_together = df[df['Order ID'].duplicated(keep =False)]
sold_together
```

```python
c = sold_together.groupby('Order ID').agg({'Product' : lambda x : ','.join(x)}).reset_index()
c
```

```python
c.set_index('Order ID')
```

```python
c.head()
```

```python
product_counts = c['Product'].value_counts()
product_counts
```

```python
c['count'] =  c['Product'].map(product_counts)
c.sort_values(by= 'count', ascending =False)
```

```python
a= c.groupby(by=c['Product']).count().reset_index()
```

```python
top5=a.sort_values(by='count',ascending=False).head(5)
top5
```

```python

sns.barplot(data=top5,y='count',x='Product')
```
